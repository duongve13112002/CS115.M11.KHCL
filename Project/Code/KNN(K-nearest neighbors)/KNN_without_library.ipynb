```python
from collections import Counter
import numpy as np
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def Manhattan_distance(x1,x2):
  return np.sum(abs(x1-x2))
```

```python
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = [self.assign_label_depend_on_neighbor(x) for x in X]
        return np.array(y_pred)

    def assign_label_depend_on_neighbor(self, x):
        # Tính khoảng cách giữa x và tất cả các examples có trong dataset
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]
        # Sắp xếp theo khoảng cách và trả về các chỉ số của k hàng xóm đầu tiên
        k_idx = np.argsort(distances)[: self.k]
        # Lấy ra các labels của KNN của các tập dataset(training)
        k_neighbor_labels = [self.y_train[i] for i in k_idx]
        # Trả về các label mà nó có nhiều giá trị nhất(most_common)
        most_common = Counter(k_neighbor_labels).most_common(1)
        return most_common[0][0]
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
def accuracy(y_true, y_pred):
  accuracy = np.sum(y_true == y_pred) / len(y_true)
  return accuracy
```

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target

```

**`Khi dùng khoảng cách là euclidean`**

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=1234
    )

k = 3
clf = KNN(k=k)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print("KNN classification accuracy", accuracy(y_test, predictions))
```
