```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
```

```python
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
print(len(np.unique(iris_y)))#In ra số nhóm của hoa 
print(len(iris_y))#Số lượng điểm dữ liệu
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris_X,iris_y,test_size = 50) #Lấy 50 data random làm test
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=10, p = 2,weights = 'distance')
#n_neighbor là số lượng điểm k gần cái diểm set nhất , p =2 là chuẩn để dựa vào nó thuộc đâu và 2 là chuẩn eclide , 1 là chuẩn mahathan và cái weights là dùng để đánh trọng số là khoảng cách giữa điểm cần xét với các điểm còn lại
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score


#Có thể tăng tỉ lệ bằng cách đánh trọng số nghĩa là khoảng cách càng xa thì trọng số càng thấp, 
print('Accuracy: %.2f%%'%(100*accuracy_score(y_test,y_pred)))#Để nâng tỉ lệ thì có thể chỉnh cái n_neighbor hoặc là dùng 
```
